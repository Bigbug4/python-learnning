```python
class Solution:
    def rob(self, nums):
        if not nums: return 0
        n=len(nums)
        if n == 1: return nums[0]
        
        dp=[0]*n
        dp[0]=nums[0]
        dp[1] = max(nums[0], nums[1])

        for i in range(2,n):
            dp[i] = max(nums[i]+dp[i-2],dp[i-1])
            #print(dp)
            
        return dp[-1]
            

a=Solution()
nums=[1,3,1,3,100]
print(nums)
m=a.rob(nums)
print(m)
```

```python
class Solution:
    def rob(self, nums):
        def robmax(nums):
            nums_len = len(nums)
            if not nums: return 0
            if nums_len == 1: return nums[0]
            
            opt = [0] * nums_len
            opt[0] = nums[0]
            opt[1] = max(nums[0],nums[1])
            
            for i in range(2,nums_len):
                opt[i] = max(opt[i-1],opt[i-2]+nums[i])
            return opt[nums_len-1]
        
        if not nums: return 0
        if len(nums) == 1: return nums[0]
        
        nums1 = nums[1:] #[1,n] 找到到最大值
        nums2 = nums[0:len(nums)-1] # [0,n-1] 找到到最大值
        return max(robmax(nums1),robmax(nums2))
            

a=Solution()
nums=[1,3,1,3,100]
print(nums)
m=a.rob(nums)
print(m)
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

# create
def create_list(nums):
    last = None
    for num in reversed(nums):
        list_node = ListNode(num)
        list_node.next = last
        last = list_node
    return last

# print 
def printList(l: ListNode):
    while l:
        print("%d, " % (l.val), end = '')
        l = l.next
    print('')


class Solution:
    def __init__(self, l1: ListNode, l2: ListNode):
        self.l1= l1
        self.l2 = l2

    def mergeTwoLists1(self,l1: ListNode, l2: ListNode) -> ListNode:
        if l1 and l2:
            if l1.val > l2.val: l1, l2 = l2, l1
            l1.next = self.mergeTwoLists1(l1.next, l2)
        return l1 or l2
        

if __name__ == "__main__":
    
    l1 = create_list([1, 2, 4])
    l2 = create_list([1, 3, 4])
    
    s = Solution(l1,l2)
    printList (s.l1)
    printList (s.l2)
```

```python
from collections import deque

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

def create_tree(values):  
    if not values: return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    leng = len(values)
    nums = 1
    while nums < leng:
        node = queue.popleft()
        if node:
            node.left = TreeNode(values[nums]) if values[nums] else None
            queue.append(node.left)
            if nums + 1 < leng:
                node.right = TreeNode(values[nums+1]) if values[nums+1] else None
                queue.append(node.right)
                nums += 1
            nums += 1
    return root

def bfs(root):
    ret = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            ret.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
    return ret

def pre_traversal(root):
    ret = []
 
    def traversal(head):
        if not head:
            return
        ret.append(head.val)
        traversal(head.left)
        traversal(head.right)
        
    traversal(root)
    return ret
 
def in_traversal(root):
    ret = []
 
    def traversal(head):
        if not head:
            return
        traversal(head.left)
        ret.append(head.val)
        traversal(head.right)
 
    traversal(root)
    return ret
 
def post_traversal(root):
    ret = []
 
    def traversal(head):
        if not head:
            return
        traversal(head.left)
        traversal(head.right)
        ret.append(head.val)
 
    traversal(root)
    return ret
    
values= [6,7,1,2,4,None,None,None,3,5,None,None,None,None,None]
root=create_tree(values)

print(bfs(root))
print(pre_traversal(root))
print(in_traversal(root))
print(post_traversal(root))
```

```python
from collections import deque

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

def create_tree(values):  
    if not values: return None
    
    root = TreeNode(values[0])
    queue = deque([root])
    leng = len(values)
    nums = 1
    while nums < leng:
        node = queue.popleft()
        if node:
            node.left = TreeNode(values[nums]) if values[nums] else None
            queue.append(node.left)
            if nums + 1 < leng:
                node.right = TreeNode(values[nums+1]) if values[nums+1] else None
                queue.append(node.right)
                nums += 1
            nums += 1
    return root

def print_tree(root):
    ret = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            ret.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
    return ret

class Solution:
    def maxDepth(self, root: TreeNode) -> int:
        if not root: return 0
        return max(self.maxDepth(root.left), self.maxDepth(root.right)) + 1
    

values= [3,2,3,None,3,None,1]

root=create_tree(values)
print(print_tree(root))

a=Solution()
m=a.maxDepth(root)
print(m)
```

```python
import  time
import copy

class solution(object):
    def __init__(self,nums):
        self.nums=nums 
        
    def insert_sort(self):
        t1=time.time()
        ilist  =  self.nums
        for i in range(len(ilist)):
            for j in range(i):
                if ilist[i] < ilist[j]:
                    ilist.insert(j, ilist.pop(i))
                    break
        t2=time.time()
        print("cost  time:  %ss"  %  (t2-t1))
        return ilist
    
    def shell_sort(self):
        t1=time.time()
        slist=self.nums
        gap = len(slist)
        while gap > 1:
            gap = gap // 2
            for i in range(gap, len(slist)):
                for j in range(i % gap, i, gap):
                    if slist[i] < slist[j]:
                        slist[i], slist[j] = slist[j], slist[i]
        
        t2=time.time()
        print("cost  time:  %ss"  %  (t2-t1))
        return slist
   
    @staticmethod
    def median(sorted_list):
        l= len (sorted_list)
        i= l % 2
        j= int((l-i)/2)
        
        if(i):
            m=sorted_list[j]
        else :
            m=(sorted_list[j-1]+sorted_list[j])/2
        
        return m
    
    def bubble_sort(self):
        blist=self.nums
        t1=time.time()
        count = len(blist)
        for j in range(0, count-1):
            for i  in range(j + 1, count-1):
                if blist[i] > blist[i+1]:
                    blist[i], blist[i+1] = blist[i+1], blist[i]
                    
        t2=time.time()
        print("cost  time:  %ss"  %  (t2-t1))
        return blist
        
    def quick_sort(self,qlist ):
        t1=time.time()
        if qlist == []:
            return []
        else:
            qfirst = qlist[0]
            qless = self.quick_sort([l for l in qlist[1:] if l < qfirst])
            qmore = self.quick_sort([m for m in qlist[1:] if m >= qfirst])
            t2=time.time()
            print("cost  time:  %ss"  %  (t2-t1))
            return qless + [qfirst] + qmore
    
    def select_sort(self):
        slist=self.nums
        t1=time.time()
        for i in range(len(slist) - 1):
            x = i
            for j in range(i, len(slist)):
                if slist[j] < slist[x]:
                    x = j
            slist[i], slist[x] = slist[x], slist[i]
            
        t2=time.time()
        print("cost  time:  %ss"  %  (t2-t1))
        return slist        
 
    def heap_sort(self):
        hlist=self.nums
        t1=time.time()
        def heap_adjust(parent):
            child = 2 * parent + 1  # left child
            while child < len(heap):
                if child + 1 < len(heap):
                    if heap[child + 1] > heap[child]:
                        child += 1  # right child
                if heap[parent] >= heap[child]:
                    break
                heap[parent], heap[child] = heap[child], heap[parent]
                parent, child = child, 2 * child + 1
 
        heap, hlist = copy.copy(hlist), []
        for i in range(len(heap) // 2, -1, -1):
            heap_adjust(i)
        while len(heap) != 0:
            heap[0], heap[-1] = heap[-1], heap[0]
            hlist.insert(0, heap.pop())
            heap_adjust(0)
        
        t2=time.time()
        print("cost  time:  %ss"  %  (t2-t1))
        return hlist
        
    def radix_sort(self):
        array=self.nums
        t1=time.time()
        bucket, digit = [[]], 0
        while len(bucket[0]) != len(array):
            bucket = [[], [], [], [], [], [], [], [], [], []]
            for i in range(len(array)):
                num = (array[i] // 10 ** digit) % 10
                bucket[num].append(array[i])
            array.clear()
            for i in range(len(bucket)):
                array += bucket[i]
            digit += 1
        
        t2=time.time()
        print("cost  time:  %ss"  %  (t2-t1)) 
        return array
    
    def merge_sort(self):
        array=self.nums
        t1=time.time()
        
        def merge_arr(arr_l, arr_r):
            array = []
            while len(arr_l) and len(arr_r):
                if arr_l[0] <= arr_r[0]:
                    array.append(arr_l.pop(0))
                elif arr_l[0] > arr_r[0]:
                    array.append(arr_r.pop(0))
            if len(arr_l) != 0:
                array += arr_l
            elif len(arr_r) != 0:
                array += arr_r
            return array
            
        def recursive(array):
            if len(array) == 1:
                return array
            mid = len(array) // 2
            arr_l = recursive(array[:mid])
            arr_r = recursive(array[mid:])
            return merge_arr(arr_l, arr_r) 
        t2=time.time()
        print("cost  time:  %ss"  %  (t2-t1)) 
        return recursive(array)  
 
if __name__ == "__main__":

    l  =  [4,5,6,7,3,2,6,9,8,11,3,8,1,13,26,21,33,18]
    
    s  =  solution(l)
    print(s.nums)
    
    import  numpy as np
    print("median:  %s\n"  %  np.median(l))
 
    ilist = s.insert_sort()
    print("insert_sort:",ilist)
    print("median:  %s\n"  %  s.median(ilist))
    
    slist = s.shell_sort()
    print ("shell_sort:",slist)
    print("median:  %s\n"  %  s.median(slist))
    
    blist = s.bubble_sort()
    print ("bubble_sort:",blist)
    print("median:  %s\n"  %  s.median(blist))
    
    qlist = s.quick_sort(l)
    print ("quick_sort:",qlist)
    print("median:  %s\n"  %  s.median(qlist))
    
    selist = s.select_sort()
    print ("select_sort:",selist)
    print("median:  %s\n"  %  s.median(selist))
    
    hlist = s.heap_sort()
    print ("heap_sort:",hlist)
    print("median:  %s\n"  %  s.median(hlist))
    
    rlist = s.radix_sort()
    print ("radix_sort:",rlist)
    print("median:  %s\n"  %  s.median(rlist))
    
    mlist = s.merge_sort()
    print ("merge_sort:",mlist)
    print("median:  %s\n"  %  s.median(mlist))
```

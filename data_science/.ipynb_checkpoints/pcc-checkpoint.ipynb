```python
# translate

text = "g fmnc wms bgblr rpylqjyrc gr zw fylb. rfyrq ufyr amknsrcpq ypc dmp. bmgle gr gl zw fylb gq glcddgagclr ylb rfyr'q ufw rfgq rcvr gq qm jmle. sqgle qrpgle.kyicrpylq() gq pcamkkclbcb. lmu ynnjw ml rfc spj."
string = [chr(i) for i in range(97,123)]
string1 = string[2:]
string1.extend(string[:2])
string = ''.join(string)
string1 = ''.join(string1)

tran = text.maketrans(string,string1)
text.translate(tran)
```

```python
# counter

with open("book.txt","r") as f:
    books = f.readlines()

books = ''.join(books)
#print(len(books))
books = books.replace("\n","")
#print(len(books))
symbol = set(books)
count = dict()

for i in symbol:
    count[i] = books.count(i)


print(count.items())
print(count.keys())
print(count.values())

L=list(count.items())       # 得到列表: L=[('a', 1), ('c', 3), ('b', 2)]
 
L.sort(key=lambda x:x[1],reverse=False)  # 按列表中，每一个元组的第二个元素从小到大排序。
                                         # x代表从L中遍历出的一个元组 
print(L)  

count_order = sorted(count.items(),key=lambda x:x[1],reverse=True)
count_order
```

```python
# candle

with open("candle.txt","r") as f:
    candle= f.read()

candle = candle.replace("\n","")   
import re
pattern = re.compile("[^A-Z][A-Z]{3}([a-z])[A-Z]{3}[^A-Z]")
results = re.findall(pattern,candle)

''.join(results)
```

```python
# follow the chain

from urllib import request

nothing = '12345'
n=4
for i in range(n):
    url = 'http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=' + str(nothing)
    req = request.urlopen(url)
    nothing = req.read().decode('utf-8')
    print(nothing)
    nothing =  nothing[-5:]  
    print('%d: %s'% (i+1,nothing))

print('\n')
print('final: %s'% nothing)
```

```python
# follow the chain

import requests
import re
import time

nothing = '12345'
i=1
while 1:
    url = 'http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=' + nothing
    req = requests.get(url)
    nothing = req.text
    print(nothing)
    result = re.findall(r'\d+', nothing)
    if result:
        nothing = result[-1]
        print('%d: %s'% (i,nothing))
        i += 1
    else:
        break
    
    time.sleep(1)

print('\n')
print('final: %s'% nothing)
```

```python
# channel

import zipfile
from urllib.request import urlretrieve


url = 'http://www.pythonchallenge.com/pc/def/channel.zip'
#urlretrieve(url, filename = 'channel.zip')
            
file_zip = zipfile.ZipFile('channel.zip')
files = file_zip.namelist()
comments = []

def item(start):

    if start in files:
        with file_zip.open(start) as f:
            content = f.read().decode('utf-8')
            print(content)
            com = file_zip.getinfo(start).comment.decode('utf-8')
            comments.append(com)
            result = '{}.txt'.format(content.split(' ').pop())
        return item(result)
    else:
        result = start.split('.')
        print('\nThe result is: %s\n' % result[0])

if __name__ == '__main__':
    item('90052.txt')
    result = ''.join(comments)
    print(result)
```

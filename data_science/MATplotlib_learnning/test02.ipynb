```python
x = np.arange(0,  3  * np.pi,  0.1) 
y = np.sin(x)
plt.title("sine wave form")  
# 使用 matplotlib 来绘制点
plt.subplot(211)
plt.plot(x, y,'r') 
plt.subplot(212)
plt.scatter(x, y)
plt.show()
```

```python
x = np.arange(0,  3  * np.pi,  0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x) 

plt.plot(x, y_sin) 
plt.plot(x, y_cos) 

plt.title('Sin & Cos')
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.legend(["sin","cos"])

plt.show()
```

```python
# 计算正弦和余弦曲线上的点的 x 和 y 坐标 
x = np.arange(0,  3  * np.pi,  0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)
y_tan = np.tan(x)
# 建立 subplot 网格，高为 2，宽为 1  
# 激活第一个 subplot
plt.subplot(1,  3,  1)
# plt.subplot(2,  1,  1)  
# 绘制第一个图像 
plt.plot(x, y_sin) 
plt.title('Sin')
plt.xlabel("x axis caption")
plt.ylabel("y axis caption") 
# 将第二个 subplot 激活，并绘制第二个图像
plt.subplot(1,  3,  2)
# plt.subplot(2,  1,  2)  
plt.plot(x, y_cos) 
plt.title('Cos')
plt.xlabel("x axis caption") 
# 将第三个 subplot 激活，并绘制第三个图像
plt.subplot(1,  3,  3)
# plt.subplot(2,  1,  2)  
plt.plot(x, y_tan) 
plt.title('Tan')
plt.xlabel("x axis caption") 
# 展示图像
plt.savefig('exe.pdf')
plt.show()
```

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,3,4*n)
y = np.linspace(-3,3,3*n)
X,Y = np.meshgrid(x,y)
plt.imshow(f(X,Y))
plt.show()
```

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.show()
```

```python
from pylab import *

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
show()
```

```python
# 构建数据
Y2016 = [15600,12700,11300,4270,3620]
Y2017 = [17400,14800,12000,5200,4020]
labels = ['北京','上海','香港','深圳','广州']
bar_width = 0.45

# 中文乱码的处理
plt.rcParams['font.sans-serif'] =['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

# 绘图
plt.bar(np.arange(5), Y2016, label = '2016', color = 'steelblue', alpha = 0.8, width = bar_width)
plt.bar(np.arange(5)+bar_width, Y2017, label = '2017', color = 'indianred', alpha = 0.8, width = bar_width)
 
plt.xlabel('Top5城市')
plt.ylabel('家庭数量')
plt.title('亿万财富家庭数Top5城市分布')
plt.xticks(np.arange(5)+bar_width,labels)
plt.ylim([2500, 19000])
 
# 为每个条形图添加数值标签
for x2016,y2016 in enumerate(Y2016):
    plt.text(x2016, y2016+100, '%s' %y2016)
 
for x2017,y2017 in enumerate(Y2017):
    plt.text(x2017+bar_width, y2017+100, '%s' %y2017)
 
plt.legend()
plt.savefig('亿万财富家庭数Top5城市分布.png')
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

show()
```
